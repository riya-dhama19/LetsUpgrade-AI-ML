## Ques1:  Python program to find the first 20 non-even prime natural numbers.

```python
print('First 20 non even Prime Numbers are :')
for num in range(3, 75):
   # all prime numbers are greater than 1
    for i in range(2, num):
        if (num % i) == 0:
            break
    else:
        print(num)
```

## Ques4. Build a User-defined function to remove all the Special Characters from string

```python
def replace_char(y):
    invalid_list = ['!','@',"#","$",'%','^','&','*','(',')','+','_','?','.',':','-'," "]
    for i in invalid_list : 
        y = y.replace(i,'')
        y = y.lower()
    print("After removing all special Characters from String")
    print("Resultant string is: ",y)
    
string_5 = 'Dr. Darshan Ingle @AI-ML Trainer'
print("String is :", string_5)
replace_char(string_5)    
```
